# Summer Olympics Data Analysis Assignment

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('summer.csv')
```

### 1. In how many cities Summer Olympics is held so far?

```python
data = []

for city in df["City"].unique():          #for avoid repeatation of cities
    data.append(city)
print("Total no of cities where olympics held was",len(data),"which are -->",data)   # Counts total no of cities
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
gold_medal = df[df["Medal"] == "Gold"]         # Selecting only Gold medal from df dataset

data = []

for sport in gold_medal["Sport"].unique():     # Selecting sport for gold medal in gold_medal dataset
    
    data.append([sport, len(gold_medal[gold_medal["Sport"] == sport])])

pd.DataFrame(data, columns = ['sport', 'Golds']).head().sort_values(by = 'Golds', ascending = False).plot(x = 'sport', y = 'Golds', kind = 'bar', figsize = (10,5), color = 'orange',title = "Summer Olympics", xlabel = 'Sport', ylabel = 'Winning Golds')
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data = []

for sport in df['Sport'].unique():     # no of times given sport name occurs is the times of when given sport wins medal.
    data.append([sport , len(df[df['Sport']  == sport])])

pd.DataFrame(data,columns = ['Sport','medals']).sort_values(by='medals', ascending=False).head().plot(x = 'Sport', y = 'medals',title = "Summer Olympics", kind = 'bar', figsize = (10,5), xlabel = 'Sport', ylabel = 'Overall medals')
```

### 4. Which player has won most number of medals? (Top 5)

```python
data = []

for player in df['Athlete'].unique():               # no of times given player name occurs is the times of when given player wins medal.
    data.append([player , len(df[df['Athlete']  == player])])

pd.DataFrame(data,columns = ['Player','medals']).sort_values(by='medals', ascending=False).head().plot(x = 'Player', y = 'medals',title = "Summer Olympics", kind = 'bar', figsize = (10,5), xlabel = 'Player', ylabel = 'Overall medals', color = "pink")
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
gold_medal = df[df["Medal"] == "Gold"]         # Selecting only Gold medal from df dataset

data = []

for athlete in gold_medal["Athlete"].unique():  # Selecting athlete for gold medal in gold_modal dataset
                                                               
    data.append([athlete, len(gold_medal[gold_medal["Athlete"] == athlete])])

pd.DataFrame(data, columns = ['Athlete', 'Golds']).head().sort_values(by = 'Golds', ascending = False).plot(x = 'Athlete', y = 'Golds', kind = 'bar', figsize = (10,5), color = 'orange',title = "Summer Olympics", xlabel = 'Athlete', ylabel = 'Winning Golds')
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
data = []
gold_medal = df[df["Medal"] == "Gold"]                                 # Getting only Gold medal data
country = gold_medal[gold_medal["Country"] == "IND"]                   # Getting only country IND
for year in country["Year"]:                                           # Fetching year from country dataset
    data.append(year)                                                  # appending
print("First Gold medal won by INDIA in Summer Olympics was in -->",data[0])  # Getting 0th index year from list
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = []

for event in df['Event'].unique():                                                         # Finding unique events
    data.append([event , len(df[df['Event']  == event])])                                  # Finding total players by length of an event

pd.DataFrame(data,columns = ['Event','Players']).sort_values(by='Players', ascending=False).head().plot(x = 'Event', y = 'Players',title = "Summer Olympics", kind = 'line', figsize = (10,5), xlabel = 'Event', ylabel = 'Total Players', color = "red")
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
gold_medal = df[df["Medal"] == "Gold"]                                           # Sorted according to Gold medal
female = gold_medal[gold_medal["Gender"] == "Women"]                             # Sorted according to female Athlets

data = []

for sport in female['Sport'].unique():                                           # counted unique sports with length
    data.append([sport , len(female[female['Sport']  == sport])])

pd.DataFrame(data,columns = ['Sport','Athlets']).sort_values(by='Athlets', ascending=False).head().plot(x = 'Sport', y = 'Athlets', kind = 'bar',title = "Summer Olympics", figsize = (10,5), xlabel = 'Sport', ylabel = 'Female Gold Medalist', color = "orange")
```
